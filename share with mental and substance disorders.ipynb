```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
from math import pi
import numpy as np
```

```python
df=pd.read_csv("share with mental and substance disorders.csv")
```

```python
df
```

```python
# Set up the matplotlib figure
plt.figure(figsize=(10, 6))

# Plotting the bar plot
plt.bar(df['Year'], df['Prevalence - Mental disorders - Sex: Both - Age: Age-standardized (Percent)'], color='skyblue')

# Set the title and labels
plt.title('Prevalence of Mental Disorders in India (1990-2022)')
plt.xlabel('Year')
plt.ylabel('Prevalence (Percent)')

# Show the plot
plt.show()
```

```python
# Convert degrees to radians for polar plot
theta = np.deg2rad(np.linspace(0, 360, len(df['Year']), endpoint=False))

# Set up the matplotlib figure in polar coordinates
plt.figure(figsize=(8, 8), subplotpars=dict(projection='polar'))

# Plotting the radial plot
plt.plot(theta, df['Prevalence - Mental disorders - Sex: Both - Age: Age-standardized (Percent)'], marker='o', label='Prevalence', color='skyblue')

# Set the title and labels
plt.title('Radial Plot of Year-wise Distribution of Mental Disorder Prevalence (1990-2022)')
plt.legend(loc='upper right')

# Show the plot
plt.show()
```
