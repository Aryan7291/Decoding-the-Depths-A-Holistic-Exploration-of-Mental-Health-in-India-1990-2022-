```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
from math import pi
import numpy as np
```

```python
df=pd.read_csv("share with mental or substance disorders by sex.csv")
```

```python
df
```

```python
# Set up the matplotlib figure
plt.figure(figsize=(10, 6))

# Plotting the line chart
plt.plot(df['Year'], df['Prevalence - Mental and substance use disorders - Sex: Male - Age: Age-standardized (Percent)'], 
         label='Male', marker='o')
plt.plot(df['Year'], df['Prevalence - Mental and substance use disorders - Sex: Female - Age: Age-standardized (Percent)'], 
         label='Female', marker='*')

# Set the title and labels
plt.title('Prevalence of Mental and Substance Use Disorders in India (1990-2020)')
plt.xlabel('Year')
plt.ylabel('Prevalence (Percent)')

# Display legend
plt.legend()

# Show the plot
plt.grid(True)
plt.show()
```
