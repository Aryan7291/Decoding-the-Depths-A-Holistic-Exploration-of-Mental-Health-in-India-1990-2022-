```python
import matplotlib.pyplot as plt
import pandas as pd
```

```python
df=pd.read_csv("share-with-depression.csv")
```

```python
df
```

```python
# Extracting relevant columns
years = df['Year']
prevalence_values = df['Prevalence - Depressive disorders - Sex: Both - Age: Age-standardized (Percent)']

# Plotting the line graph
plt.figure(figsize=(10, 6))
plt.plot(years, prevalence_values, marker='o', linestyle='-')

# Adding labels and title
plt.title('Prevalence of Depressive Disorders in India (1990-2022)')
plt.xlabel('Year')
plt.ylabel('Prevalence (Percent)')

# Display the graph
plt.grid(True)
plt.show()
```

```python
plt.figure(figsize=(12, 6))
plt.plot(df['Year'], df['Prevalence - Depressive disorders - Sex: Both - Age: Age-standardized (Percent)'], 
marker='o', linestyle='-', color='b')
plt.title('Prevalence of Depressive Disorders in India (1990-2022)')
plt.xlabel('Year')
plt.ylabel('Prevalence (Percent)')
plt.grid(True)
plt.show()
plt.savefig('Prevalence of Depressive Disorders in India (1990-2022)')
```

```python
# Visualization 4: Histogram
plt.figure(figsize=(8, 6))
plt.hist(df['Prevalence - Depressive disorders - Sex: Both - Age: Age-standardized (Percent)'], bins=10, color='lightcoral', edgecolor='black')
plt.title('Histogram of Depressive Disorders Prevalence in India (1990-2022)')
plt.xlabel('Prevalence (Percent)')
plt.ylabel('Frequency')
plt.grid(axis='y')
plt.show()
```

```python
# Visualization 5: Area Plot
plt.figure(figsize=(12, 6))
plt.fill_between(df['Year'], df['Prevalence - Depressive disorders - Sex: Both - Age: Age-standardized (Percent)'], color='skyblue', alpha=0.4)
plt.plot(df['Year'], df['Prevalence - Depressive disorders - Sex: Both - Age: Age-standardized (Percent)'], marker='', linestyle='-', color='b')
plt.title('Area Plot of Depressive Disorders Prevalence in India (1990-2022)')
plt.xlabel('Year')
plt.ylabel('Prevalence (Percent)')
plt.grid(True)
plt.show()


```
