```python
import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
df=pd.read_csv("mental-and-substance-use-as-share-of-disease.csv")
```

```python
df
```

```python
# Plotting
plt.figure(figsize=(10, 6))
plt.plot(df['Year'], df['DALYs (Disability-Adjusted Life Years) - Mental disorders - Sex: Both - Age: All Ages (Percent)'], marker='o', linestyle='-', color='b')
plt.title('Disability-Adjusted Life Years (DALYs) for Mental Disorders in India (1990-2019)')
plt.xlabel('Year')
plt.ylabel('DALYs (Percent)')
plt.grid(True)
plt.show()
```

```python
# Reshape the data for heatmap
heatmap_data = df.pivot(index='Year', columns='Code', values=
'DALYs (Disability-Adjusted Life Years) - Mental disorders - Sex: Both - Age: All Ages (Percent)')
# Set up the matplotlib figure
plt.figure(figsize=(12, 8))
# Create the heatmap
sns.heatmap(heatmap_data, cmap='YlGnBu', annot=True, fmt=".3f", linewidths=.5)
# Set the title and show the plot
plt.title('Heatmap of DALYs for Mental Disorders in India (1990-2022)')
plt.show()
```
