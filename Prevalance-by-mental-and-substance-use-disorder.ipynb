```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
from math import pi
```

```python
df=pd.read_csv('Prevalance-by-mental-and-substance-use-disorder.csv')
```

```python
df
```

```python
# Set up the matplotlib figure
plt.figure(figsize=(10, 6))

# Create a scatter plot
sns.scatterplot(x='Prevalence - Schizophrenia - Sex: Both - Age: Age-standardized (Percent)', 
                y='Prevalence - Bipolar disorder - Sex: Both - Age: Age-standardized (Percent)', data=df)

# Set the title and labels
plt.title('Scatter Plot of Schizophrenia vs. Bipolar Disorder Prevalence (1990-2022)')
plt.xlabel('Prevalence - Schizophrenia (Percent)')
plt.ylabel('Prevalence - Bipolar Disorder (Percent)')

# Show the plot
plt.show()
```

```python
# Set up the matplotlib figure
plt.figure(figsize=(10, 6))

# Plotting the comparative line chart
sns.lineplot(x='Year', y='Prevalence - Alcohol use disorders - Sex: Both - Age: Age-standardized (Percent)', data=df, label='Alcohol Usage Disorder', marker='o')
sns.lineplot(x='Year', y='Prevalence - Drug use disorders - Sex: Both - Age: Age-standardized (Percent)', data=df, label='Drug use Disorder', marker='o')

# Set the title and labels
plt.title('Comparative Line Chart of Alcohol Usage Disorder vs. Drug use Disorder (1990-2022)')
plt.xlabel('Year')
plt.ylabel('Prevalence (Percent)')

# Display legend
plt.legend()

# Show the plot
plt.show()
```
