```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
from math import pi
```

```python
df=pd.read_csv("prevalance-of-depression-males-vs-females.csv")
```

```python
df
```

```python
# Set up the matplotlib figure
plt.figure(figsize=(10, 6))

# Plotting the comparative line chart
sns.lineplot(x='Year', y='Prevalence - Depressive disorders - Sex: Male - Age: Age-standardized (Percent)', 
             data=df, label='Male', marker='o')
sns.lineplot(x='Year', y='Prevalence - Depressive disorders - Sex: Female - Age: Age-standardized (Percent)', 
             data=df, label='Female', marker='*')

# Set the title and labels
plt.title('Comparative Line Chart of Male vs. Female Prevalence of Depressive Disorders (1990-2022)')
plt.xlabel('Year')
plt.ylabel('Prevalence (Percent)')

# Display legend
plt.legend()

# Show the plot
plt.show()
```
